```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('/content/drive/MyDrive/Files/bmi.csv')

# Extract features and target variable as dataframes, not lists
x = data[['Height','Weight']]
y = data['Index']

LR = LinearRegression()

LR.fit(x,y)
```

```python
user_Height = float(input("Enter Height: "))
user_Weight = float(input("Enter Weight: "))

user_input = pd.DataFrame({'Height': [user_Height], 'Weight': [user_Weight]})
result = LR.predict(user_input)
print('BMI =', result)

# Round the result to the nearest integer for comparison
result = round(result[0])

if result == 0:
    print('Extremely Weak')
elif result == 1:
    print('Weak')
elif result == 2:
    print('Normal')
elif result == 3:
    print('Overweight')
elif result == 4:
    print('Obesity')
elif result == 5:
    print('Extreme Obesity')
else:
    print('Result out of bounds')
```
